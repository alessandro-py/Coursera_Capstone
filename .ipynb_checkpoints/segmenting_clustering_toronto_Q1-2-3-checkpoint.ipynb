```python
#QUESTION 1
from sklearn.cluster import KMeans
import json
import requests
import bs4 as bs
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
#importing the url data into text
url ='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
url
results = requests.get(url).text
```

```python
#parsing the data with BeautifulSoup into lxml
DataToronto = BeautifulSoup(results, 'lxml')
#finding the table part of the data
table=DataToronto.find('table')
#Assigning column to the df
column_names = ['PostalCode','Borough','Neighborhood']
df = pd.DataFrame(columns=column_names)
df
```

```python
#assign all 'tr' and 'td' data to rows 
for tr_cell in table.find_all('tr'):
    row_data=[]
    for td_cell in tr_cell.find_all('td'):
        row_data.append(td_cell.text.strip())
    if len(row_data)==3:
        df.loc[len(df)] = row_data
#drop all the data the data that is not assigned
df.drop(df[df['Borough']=="Not assigned"].index,axis=0, inplace=True)
#reset index, Sorting by PostalCode and Erasing Duplicates
df= df.sort_values(by=['PostalCode'])
df= df.drop_duplicates(subset='PostalCode')
df =df.reset_index(drop=True)
#diplay the whole df
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    df.head()
    
```

```python
# Code to Hide Client ID, Secret and Access token
from IPython.display import HTML
import random

def hide_toggle(for_next=False):
    this_cell = """$('div.cell.code_cell.rendered.selected')"""
    next_cell = this_cell + '.next()'

    toggle_text = 'Toggle show/hide'  # text shown on toggle link
    target_cell = this_cell  # target cell to control with toggle
    js_hide_current = ''  # bit of JS to permanently hide code in current cell (only when toggling next cell)

    if for_next:
        target_cell = next_cell
        toggle_text += ' next cell'
        js_hide_current = this_cell + '.find("div.input").hide();'

    js_f_name = 'code_toggle_{}'.format(str(random.randint(1,2**64)))

    html = """
        <script>
            function {f_name}() {{
                {cell_selector}.find('div.input').toggle();
            }}

            {js_hide_current}
        </script>

        <a href="javascript:{f_name}()">{toggle_text}</a>
    """.format(
        f_name=js_f_name,
        cell_selector=target_cell,
        js_hide_current=js_hide_current, 
        toggle_text=toggle_text
    )

    return HTML(html)
hide_toggle(for_next=True)
```

```python
toronto_venues = getNearbyVenues(names= df['Neighborhood'],
                                   latitudes= df['Latitude'],
                                   longitudes= df['Longitude'])
```

```python
#dividing all the food places from other venues so that we can create a cluster
result1= toronto_venues[toronto_venues['Venue Category'].str.contains 
                        ('Restaurant|Place|Joint|Bodega|Diner|Pie Shop|Fish & Chips Shop|Food|Brasserie')]
```

```python
lat_long_data= result1.drop(columns=['Neighborhood Latitude','Neighborhood Longitude','Venue'])
```

```python
lat_long_data.groupby('Venue Category').count()
```

```python
# one hot encoding
toronto_onehot = pd.get_dummies(lat_long_data[['Neighborhood']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Venue Category'] = result1['Venue Category'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]
```

```python
toronto_grouped = toronto_onehot.groupby('Venue Category').mean().reset_index()
toronto_grouped.head()
```

```python
toronto_venues.groupby('Neighborhood').count()


# one hot encoding
toronto_onehot1 = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")
toronto_onehot1.drop(['Neighborhood'],axis=1,inplace=True) 
toronto_onehot1.insert(loc=0, column='Neighborhood', value=toronto_venues['Neighborhood'] )
toronto_onehot1.shape



toronto_grouped1 = toronto_onehot1.groupby('Neighborhood').mean().reset_index()
toronto_grouped1.head()

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
toronto_grouped1.head()
```

```python
num_top_venues = 25

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Venue Category']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Venues'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Venues'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Venue Category'] = toronto_grouped['Venue Category']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

toronto_grouped.head()
```

```python
num_top_venues = 5

for hood in toronto_grouped['Venue Category']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Venue Category'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Venue Category', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)
# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:20] 
```

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = lat_long_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Venue Category'), on='Venue Category')
```
